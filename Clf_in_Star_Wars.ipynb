```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
test_data = pd.read_csv('invasion.csv')
test_data
```

```python
X_train = test_data.drop(['class'], axis = 1)
y_train = test_data['class']
```

```python
clf = DecisionTreeClassifier()
```

```python
params = {'max_depth' : range(1,8), 'min_samples_split': range(1,13,2), 'min_samples_leaf': range(2,10,2)}
search_grid = GridSearchCV(clf, params, cv = 5)
```

```python
search_grid.fit(X_train, y_train)
```

```python
search_grid.best_params_
```

```python
test= pd.read_csv('operative_information.csv')
```

```python
test
```

```python
best_tree =  DecisionTreeClassifier(max_depth = 2, min_samples_leaf = 2, min_samples_split = 3)
best_tree.fit(X_train, y_train)
y_pred_test = best_tree.predict(test)
```
